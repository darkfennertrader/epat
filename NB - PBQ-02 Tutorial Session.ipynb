## Modifying a dataframe

##### Adding new columns

````python
# Example 17
# Adding a new column to the dataframe and assigning a static value
data['returns'] = 0

data.head()
````

````python
# Counting the values in a column - Using value counts
data['returns'].value_counts()
````

````python
# Example 18
# Adding a new column to the dataframe and assigning dynamic value

# E-18A - Calculate percentage returns
data['percentage_change'] = data['Adj Close'].pct_change()

data.head()
````

````python
# E-18B - Convert prices to log prices np.log()
data['log_returns'] = np.log(data['Adj Close'] / data['Adj Close'].shift(1))

data.head(10)

# E-18C - Generate log returns np.log(today's price / yesterday's price)
````

````python
# Example 19
# Adding a new column to the dataframe and assigning values based on a logical condition
positive_percent = data['percentage_change'] > 0 # 1
negative_percent = data['percentage_change'] < 0 # -1

data['movement'] = np.where(positive_percent, 1, np.where(negative_percent, -1, 0))

data.head()
````

````python
# Another example
buy_condition = data['movement'] == 1
sell_condition = data['movement'] == -1

data['string'] = np.where(buy_condition, 'Buy', np.where(sell_condition, 'Sell', 'No Change'))

data.head()
````

````python
# Counting the values in a column
data['movement'].value_counts()
````

````python
# Plotting it
data['movement'][:100].plot(figsize=(12, 4));
````

````python
# To install plotly library
# Use version 4.14.3
!pip install plotly==4.14.3
````

````python
# Example 26
# Import plotly library
import plotly.graph_objs as go
````

````python
data = pd.read_csv('AAPL_Daily_Data.csv', index_col=0, parse_dates=True)
````

#### Method 1

````python
# Create chart object
chart_data = [go.Candlestick(x=data.index[:100], 
                             open=data['Open'][:100], 
                             high=data['High'][:100], 
                             low=data['Low'][:100],
                             close=data['Close'][:100]
                            )]

# Load chart data
fig = go.Figure(data=chart_data)

# Update chart layout
fig.update_layout(xaxis_rangeslider_visible=False, xaxis_showticklabels=True, yaxis_showticklabels=True)

# Plot chart
fig.show()
````

````python
import cufflinks as cf
cf.go_offline()

# Plot candlestick chart
qf = cf.QuantFig(data, title='AAPL Stock Price', legend='top')
qf.add_volume()
qf.add_sma(periods=20, column='Close', color='red')
qf.add_ema(periods=20, column='Close', color='green')
qf.iplot()
````
